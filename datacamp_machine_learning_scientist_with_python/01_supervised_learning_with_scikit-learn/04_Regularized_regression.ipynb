# Why regularize?

Regression optimizes the loss function to choose a coefficient $a_i$. If the coefficients are too large, we risk overfitting the model. Regularization penalizes large coefficients. There are two types of regularization:

## L1 regularization

It adds a penalty that is equal to the absolute value of the magnitude of the coefficients. Some of the coefficients can be reduced to zero and thus eliminated, so L1-type regularization can result in models with a lower number of cofficients. 

Examples: Lasso regularization

## L2 regularization

It adds a penalty equal to the square of the coefficients. It doesn't reduce the amount of coefficients: they are all reduced by the same factor and none is eliminated.

Examples: Ridge regression and SVMs.

__We always use Ridge first, then Lasso.__

### Elastic nets: 

They combine L1 and L2, but adding a hypterparameter.
